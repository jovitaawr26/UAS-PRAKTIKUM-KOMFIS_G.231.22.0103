a.) Distribusi data pada setiap atribut Purchase Price, Current Price dan Gain/Loss dengan
grafik histogram :

```python
import matplotlib.pyplot as plt
import pandas as pd

# Simulasi data
data = {
    'Stock Name': ['Stock A', 'Stock B', 'Stock C', 'Stock D', 'Stock E'],
    'Symbol': ['A', 'B', 'C', 'D', 'E'],
    'Purchase Price': [100, 150, 200, 120, 180],
    'Current Price': [120, 160, 220, 110, 170],
    'Gain/Loss': [20, 10, 20, -10, -10],
    'Dividen / Share': [5, 8, 3, 6, 4],
    'Market Value': [500, 800, 350, 600, 400],
    'Annual Yield': [0.05, 0.06, 0.03, 0.04, 0.02]
}

df = pd.DataFrame(data)

# a. Histogram untuk atribut Purchase Price, Current Price, dan Gain/Loss
plt.figure(figsize=(10, 6))

plt.subplot(2, 2, 1)
plt.hist(df['Purchase Price'], bins=10, color='skyblue')
plt.xlabel('Purchase Price')
plt.ylabel('Count')
plt.title('Distribution of Purchase Price')

plt.subplot(2, 2, 2)
plt.hist(df['Current Price'], bins=10, color='lightgreen')
plt.xlabel('Current Price')
plt.ylabel('Count')
plt.title('Distribution of Current Price')

plt.subplot(2, 2, 3)
plt.hist(df['Gain/Loss'], bins=10, color='orange')
plt.xlabel('Gain/Loss')
plt.ylabel('Count')
plt.title('Distribution of Gain/Loss')

plt.tight_layout()
plt.show()


```

c. Atribut Stock Name dan Symbol yang memiliki Market Value 5 tertinggi

```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca file CSV
data = pd.read_csv('stock1.csv')

# Mengurutkan berdasarkan Market Value secara menurun
sorted_data = data.sort_values('Market Value', ascending=False)

# Mengambil 5 data teratas
top_market_value = sorted_data.head(5)[['Stock Name', 'Symbol']]

# Menampilkan tabel dengan matplotlib
fig, ax = plt.subplots(figsize=(6, 4))
ax.axis('off')  # Menghilangkan sumbu

table = ax.table(cellText=top_market_value.values, colLabels=top_market_value.columns, loc='center', cellLoc='center')
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1, 1.5)  # Menyesuaikan skala tabel

plt.title('Top 5 Stocks by Market Value')

plt.show()
```
